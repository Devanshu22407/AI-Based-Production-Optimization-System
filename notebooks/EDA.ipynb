# EDA - Oncology Production Batch Dataset
This notebook explores synthetic production data and key analytics for processing time and resource allocation.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

data_path = Path('../backend/data/production_data.csv')
df = pd.read_csv(data_path)
df.head()
```

```python
df.describe(include='all').T
```

```python
plt.figure(figsize=(8, 4))
df['Processing_Time'].hist(bins=20)
plt.title('Processing Time Distribution')
plt.xlabel('Minutes')
plt.ylabel('Frequency')
plt.show()
```

```python
machine_util = df.groupby('Machine_ID')['Machine_Load'].mean().sort_values(ascending=False) * 100
machine_util
```
